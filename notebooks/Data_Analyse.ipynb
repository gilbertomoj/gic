```python
import pandas as pd
year = "2019"
# Leitura dos dados
df = pd.read_csv(f"../../sisu_data/{year}/chamada_regular_{year}-1.csv", encoding='UTF-8', on_bad_lines='skip', sep=";")

df
```

```python
# Listagem das faculdades no banco de dados
df["NO_IES"].unique()
```

```python
# Selecionando os registros apenas do UFPE
pe = df.loc[df["NO_IES"] == "UNIVERSIDADE FEDERAL DE PERNAMBUCO"][df["NO_MUNUCIPIO_CAMPUS"] == "Recife"]
pe
```

```python
new_title = "pe_data"

# Escrevendo a tabela filtrada e padronizando o encoding para UTF-8
pe.to_csv(f"../{year}/{new_title}_{year}-1.csv", sep="|", index=False, encoding='UTF-8')
```
